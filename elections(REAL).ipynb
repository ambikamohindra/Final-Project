```
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('precision', 2)
```

```
# Reading the CSV File
df = pd.read_csv('2020_elections.csv')
```

```
# Drop the columns that are not needed
df = df.drop(['Vote for Highest Office (President)', 'Overseas Eligible'], axis = 1)
df = df.drop(0, axis=0) 
```

```
# Average percentage of the state that constitutes non-citizen
# HINT: Use a statistic method on the columns
df['% Non-citizen'] = [float(stat[:-1]) for stat in list(df['% Non-citizen'])]
df['VEP Turnout Rate'] = [float(stat[:-1]) for stat in list(df['VEP Turnout Rate'])]
```

```
df.head()
```

```
def change_type(name):
  items = list(df[name])
  for i in range(len(items)):
    items[i] = float(items[i].replace(',' , ''))
  df[name] = items
```

```
change_type('Total Ballots Counted (Estimate)')
change_type('Voting-Eligible Population (VEP)')
change_type('Voting-Age Population (VAP)')
change_type('Prison')
change_type('Probation')
change_type('Parole')
change_type('Total Ineligible Felon')
```

```
import plotly.express as px
fig = px.scatter(df, x="Voting-Eligible Population (VEP)", y="Total Ineligible Felon", hover_name='State Abv', 
              title='Voting Eligible Population vs Total Ineligible Population', trendline="ols")
fig.show()
```

```
top = ['california', 'arizona' 'tennessee', 'kansas', 'oklahoma', 'arkansas', 'rhode island', 'utah', 'new york', 'alabama']

most_covid = []
for state in list(df['State']):
    if state.lower() in top:
        most_covid.append('Top 10 Highest Covid per Capita')
    else:
        most_covid.append('Not Top Covid State')
        
df['covid cases'] = most_covid
```
